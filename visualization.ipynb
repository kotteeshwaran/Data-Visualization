```python
%matplotlib inline
import matplotlib.pyplot as plt

fig,axis = plt.subplots()
axis.yaxis.grid(True)
axis.set_title('CO2 Emission vs. GDP')
axis.set_xlabel(gdp_stage_trim['IndicatorName'].iloc[0])
axis.set_ylabel(stage['IndicatorName'].iloc[0])

x= gdp_stage_trim['Value']
y= stage['Value']

axis.scatter(x,y)
plt.show()
```
